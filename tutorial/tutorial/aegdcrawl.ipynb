```python
programs = []
with open('../../aegd.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        programs.append(row)
    
```

```python

for value in programs[]:
    for j in search(value["Program Name"] + " " + value["Program Type"], tld="co.in", num=1, stop=1, pause=2):
        print(value["Program Name"])
        value["url"] = j
    
```

```python
with open("../../aegd_output.csv", "w") as csv_writer:
    writer = csv.DictWriter(csv_writer, fieldnames=["Program Name", "Program Type", "url"])
    writer.writeheader()
    writer.writerows(programs)
```

```python
import spacy
from collections import Counter
from string import punctuation
nlp = spacy.load("en_core_web_sm")
from spacy.matcher import Matcher


def get_hotwords(text):
    result = []
    pattern = [{'POS': 'ADJ', 'DEP': 'amod'},
               {'POS': 'NOUN'}]
    matcher = Matcher(nlp.vocab)
    matcher.add("DentalVocab", [pattern])
    doc = nlp(text.lower())
    matches = matcher(doc)
    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]  # Get string representation
        span = doc[start:end]  # The matched span
        print(match_id, string_id, start, end, span.text)
        result.append(span.text)
    
    return result

with open("keywords.txt", "w") as write_file:
    total = set()
    for doc in docs:
        text = doc["body"] if "body" in doc else ""
        output = set(get_hotwords(text))
        for line in output:
            total.add(line)
    for line in total:
        write_file.write(line + "\n")
```
